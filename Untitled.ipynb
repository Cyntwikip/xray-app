```python
import base64 
with open("images/neko_girl_1.jpg", "rb") as f:
    data = f.read()
#     encoded_string = base64.encodestring(data)
    encoded_string = base64.b64encode(data)
```

```python
import matplotlib.image as mpimg
# plt.imshow(mpimg.imread(data))
```

```python
encoded_string
```

```python
# base64.decodestring(encoded_string)
# encoded_string.decode('base64')
img2 = base64.b64decode(encoded_string)
```

```python
with open('images/test.jpg', 'wb') as f:
    f.write(img2)
```

```python
import io 

i = io.BytesIO(img2)
i = mpimg.imread(i, format='JPG')

# plt.imshow(i, interpolation='nearest')
plt.imshow(i)
plt.show()
```

```python
from io import BytesIO

def convertToJpeg(im):
    with BytesIO() as f:
        im.save(f, format='PNG')
        return f.getvalue()
    
img3 = Image.fromarray(i)
img3 = convertToJpeg(img3)
base64.b64encode(img3)
```
